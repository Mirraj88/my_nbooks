```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('avocado.csv')
df.head(5)
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df.query('year == 2015 & type == "conventional" & region == "Albany"')[['Date','Total Bags','Small Bags']].head(5)
```

```python
melt_df = pd.melt(df, 
                  id_vars = 'Date',
                  value_vars = ['Total Bags', 'Small Bags'],
                  var_name = 'Bags',
                  value_name = 'Count')
melt_df
```

```python
sns.relplot(data = melt_df, kind = 'line', x = 'Date', y = 'Count', hue = 'Bags')
```

```python
df['Sales Volume'] = pd.qcut(
                        df['Total Volume'],
                        q = 4,
                        labels = ['Poor', 'Modest', 'Good', 'Excellent']
                        )
df.head(5)
```

```python
sns.countplot(data = df, x = df['year'], hue = 'Sales Volume')
```

```python
melt_df = pd.melt(df,
                  id_vars = 'year',
                  value_vars = ['Small Bags', 'Large Bags', 'XLarge Bags'],
                  var_name = 'Bags',
                  value_name = 'Count')
melt_df
```

```python
sns.relplot(data = melt_df, kind = 'line', x = 'year', y = 'Count', hue = 'Bags')
```
